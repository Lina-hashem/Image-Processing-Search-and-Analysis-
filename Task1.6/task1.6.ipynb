```python
import numpy as np
import imageio
# from PIL import Image
import matplotlib.pyplot as plt
```

Read photo

```python
arrF = imageio.imread('portrait.png', pilmode='L')
result_file_name='result_1.6.png'
```

Add white ellipse

```python
M, N = arrF.shape
c = np.array(arrF.shape) / 2
L = 80
W = 45
rs, cs = np.meshgrid(np.arange(M), np.arange(N), indexing='ij')
ff = ((rs - c[0]) / W) ** 2
gg = ((cs - c[1]) / L) ** 2
indexes = np.where(ff + gg <= 1)
arrF[indexes] = 255
imageio.imwrite(result_file_name, arrF)
```

Show the result

```python

result=imageio.imread(result_file_name,pilmode='RGB')
plt.imshow(result / 255)
plt.show()
```
