```python
import numpy as np
import imageio as iio
import matplotlib.pyplot as plt
```

```python
result_image_file_name='result_1.5.png'
src_image_file_name='portrait.png'
```

```python
arrF = iio.imread(src_image_file_name,pilmode='L')
h,w=arrF.shape  
arrF[0::16,]=255
arrF[:,0::16]=255
iio.imwrite(result_image_file_name,arrF)   
result_image = iio.imread(result_image_file_name,pilmode='RGB')
plt.imshow(result_image / 255)
plt.show()
```
