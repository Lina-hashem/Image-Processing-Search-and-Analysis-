```python
import imageio
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
```

```python
def imageRead(imgname, pilmode='L', arrtype=np.float):
    return imageio.imread(imgname, pilmode=pilmode).astype(arrtype)
def imageWrite(arrF, imgname, arrtype=np.uint8):
    imageio.imwrite(imgname, arrF.astype(arrtype))
```

```python
options=[(4, 4), (32, 32), (16, 32), (32, 16) ]
```

```python
def meanSuperPixelV2(arrF, m, n):
    M, N = arrF.shape
    arrG = np.zeros((M, N))
    for i in range(0, M, m):
        for j in range(0, N, n):
            arrG[i:i+m,j:j+n] = np.mean(arrF[i:i+m,j:j+n])
    return arrG
```

```python
def medianSuperPixel(arrF,m,n):
    M, N = arrF.shape
    arrG = np.zeros((M, N))
    for i in range(0, M, m):
        for j in range(0, N, n):
            arrG[i:i+m,j:j+n] = np.mean(arrF[i:i+m,j:j+n])
    return arrG
```

```python
arrF=imageRead('portrait.png')
for counter,(a,b) in enumerate(options):
    arrG=medianSuperPixel(arrF=arrF,m=a,n=b)
    imageWrite(arrF=arrG,imgname="medianSuperPixel{}.png".format(counter))    
for counter,(a,b) in enumerate(options):
    arrG2=meanSuperPixelV2(arrF=arrF,m=a,n=b)
    imageWrite(arrF=arrG,imgname="meanSuperPixelV2{}.png".format(counter))    
```
